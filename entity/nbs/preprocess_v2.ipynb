```python
import json
import pandas as pd
```

```python
data = json.load(open('../../data/ner/NER_TRAIN/NER_TRAIN_JUDGEMENT.json'))
```

```python
text = data[0]['data']['text']
annot = data[0]['annotations'][0]['result']
```

```python
words = []

for w in text.split(' '):
    words.extend(w.split('\n'))

for i, w in enumerate(words):
    if i == len(words) - 1:
        continue
    
    if words[i + 1] == '':
        words[i] = words[i] + '\n'
        words[i + 1] = '\n'
        # words[i + 1] = 'z'


for i, w in enumerate(words):
    try:
        if words[i][-1] != '\n':
            words[i] += ' '
    except:
        pass

for i, w in enumerate(words):
    try:
        if words[i][-1] == '\n' and len(words[i]) != 1:
            words.insert(i + 1, '\n')
    except:
        pass
```

```python
text[73:83]
```

```python
tags = [0 for _ in range(len(text))]

for i, w in enumerate(text):
    for x in annot:
        start, end, label = x['value']['start'], x['value']['end'], x['value']['labels'][0]
        if i >= start and i <= end:
            tags[i] = label
            break
```

```python
tags
```

```python
for w, t in zip(text, tags):
    print(ascii(w), t)
```

```python
words, final_tags = [], []

i = 0

while i < len(tags):
    if tags[i] == 0:
        w = ''
        while i < len(text) and text[i] != ' ' and tags[i] == 0:
            w += text[i]
            i += 1
        # print(1, w)
        words.append(w)
        final_tags.append('OBJECT')
        i += 1
    else:
        w = ''
        curr_tag = tags[i]
        while (i < len(tags) - 1) and (tags[i] == tags[i + 1]) and tags[i] != 0:
            w += text[i]
            i += 1
        # print(2, w)
        words.append(w)
        final_tags.append(curr_tag)

        if tags[i] != curr_tag:
            pass
        else:
            i += 1
```

```python
for w, t in zip(words, final_tags):
    if t != 'OBJECT':
        print(w, t)
```

```python
import pandas as pd

df = pd.DataFrame({
    'words': words,
    'tags': final_tags
})
```
